```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# --- Step 1: Data Loading (Placeholder - replace with your data) ---
# Assuming your data is in a CSV file. Replace 'your_data.csv' with your file path.
# df = pd.read_csv('your_data.csv')

# --- Dummy DataFrame for demonstration purposes ---
# This part is for demonstration. In a real scenario, you'd load your own data.
np.random.seed(42)
data = {
    'danceability': np.random.rand(100) * 0.8 + 0.2,
    'energy': np.random.rand(100) * 0.8 + 0.2,
    'loudness': np.random.rand(100) * 20 - 40,
    'speechiness': np.random.rand(100) * 0.2,
    'acousticness': np.random.rand(100) * 0.5,
    'instrumentalness': np.random.rand(100) * 0.5,
    'liveness': np.random.rand(100) * 0.4,
    'valence': np.random.rand(100) * 0.8 + 0.2,
    'tempo': np.random.rand(100) * 100 + 80,
    'track_artist': [f'Artist_{i % 10}' for i in range(100)],
    'playlist_genre': ['pop', 'rock', 'hip-hop', 'electronic'] * 25
}
df = pd.DataFrame(data)

# --- Step 2: Feature Selection and Scaling ---
# Select features for clustering.
features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
            'instrumentalness', 'liveness', 'valence', 'tempo']

# Create a new DataFrame with just these features
clustering_data = df[features]

# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# --- Step 3: Elbow Method to Determine Optimal K ---
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.grid(True)
plt.show()

# --- Step 4: Fit K-Means Model and Assign Clusters ---
# Based on the elbow plot, let's choose K=4 as a reasonable number of clusters.
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(scaled_data)

# --- Step 5: Visualize Clusters with PCA ---
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['cluster'] = df['cluster']

plt.figure(figsize=(12, 8))
sns.scatterplot(x='PC1', y='PC2', hue='cluster', data=pca_df, palette='viridis', style='cluster', s=100)
plt.title('Clusters of Songs (PCA-reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# --- Step 6: Analyze Cluster Characteristics ---
print("\nCluster Characteristics (Average values of features):")
cluster_characteristics = df.groupby('cluster')[features].mean()
print(cluster_characteristics)

# Find top artists and genres in each cluster
for i in sorted(df['cluster'].unique()):
    print(f"\n--- Cluster {i} ---")
    cluster_df = df[df['cluster'] == i]
    top_genres = cluster_df['playlist_genre'].value_counts().nlargest(3)
    top_artists = cluster_df['track_artist'].value_counts().nlargest(3)
    print("Top Genres:\n", top_genres)
    print("Top Artists:\n", top_artists)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler

# --- Step 1: Data Loading (Placeholder - replace with your data) ---
# Assuming your data is in a DataFrame named 'df'.
# For demonstration, a dummy DataFrame is created.
np.random.seed(42)
data = {
    'track_name': [f'Song_{i}' for i in range(10000)],
    'track_artist': [f'Artist_{i % 500}' for i in range(10000)],
    'playlist_genre': ['pop', 'rock', 'hip-hop', 'electronic', 'jazz'] * 2000,
    'danceability': np.random.rand(10000) * 0.8 + 0.2,
    'energy': np.random.rand(10000) * 0.8 + 0.2,
    'loudness': np.random.rand(10000) * 20 - 40,
    'speechiness': np.random.rand(10000) * 0.2,
    'acousticness': np.random.rand(10000) * 0.5,
    'instrumentalness': np.random.rand(10000) * 0.5,
    'liveness': np.random.rand(10000) * 0.4,
    'valence': np.random.rand(10000) * 0.8 + 0.2,
    'tempo': np.random.rand(10000) * 100 + 80
}
df = pd.DataFrame(data)

# --- Step 2: Feature Selection and Scaling ---
features_for_recomm = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                       'instrumentalness', 'liveness', 'valence', 'tempo']

# Let's re-scale the features using MinMaxScaler
scaler = MinMaxScaler()
df[features_for_recomm] = scaler.fit_transform(df[features_for_recomm])

# --- Step 3: Compute Cosine Similarity ---
# For a large dataset, we must use a sample to avoid memory issues.
# We will use the sample to build our recommendation system.
sample_df = df.sample(n=5000, random_state=42).copy()
normalized_sample_features = sample_df[features_for_recomm].values
cosine_sim = cosine_similarity(normalized_sample_features)

# Create a mapping from song name to index
indices = pd.Series(sample_df.index, index=sample_df['track_name']).drop_duplicates()

# --- Step 4: Recommendation Function ---
def get_recommendations(track_name, cosine_sim=cosine_sim, df=sample_df, indices=indices):
    """
    Generates song recommendations based on content similarity.
    """
    # Get the index of the song from the sample DataFrame that matches the track name
    if track_name not in indices:
        return f"'{track_name}' not found in the dataset. Please try a different song."

    # We need to get the index within the *sample* DataFrame.
    # The `loc` method of the index object returns the index label, which is what we need.
    idx_in_sample_df = indices.loc[track_name]

    # Get the row corresponding to this index from the sample DataFrame to find its position
    # in the cosine_sim matrix.
    position_in_matrix = df.index.get_loc(idx_in_sample_df)

    # Get the pairwise similarity scores of all songs with that song from the matrix.
    sim_scores = list(enumerate(cosine_sim[position_in_matrix]))

    # Sort the songs based on the similarity scores.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 11 most similar songs (including the song itself).
    sim_scores = sim_scores[0:11]

    # Get the song indices from the sample DataFrame.
    song_indices_in_sample = [i[0] for i in sim_scores]

    # Get the actual index labels from the sample DataFrame.
    actual_indices = df.iloc[song_indices_in_sample].index.tolist()

    # Remove the input song from the recommendations.
    try:
        actual_indices.remove(idx_in_sample_df)
    except ValueError:
        pass  # The song might not be in the top 10 for some reason.

    # Get the top 10 most similar songs.
    recommendations_df = df.loc[actual_indices].iloc[0:10][['track_name', 'track_artist', 'playlist_genre', 'danceability', 'energy', 'valence']]

    return recommendations_df

# --- Example Usage ---
# Use a song name from the dummy dataset
song_to_recommend = "Song_123"
recommendations = get_recommendations(song_to_recommend)

print(f"--- Top 10 recommendations for '{song_to_recommend}' ---")
print(recommendations)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler

# --- Step 1: Data Loading (Placeholder) ---
# Create a dummy DataFrame for demonstration. In a real scenario, you'd load your data.
np.random.seed(42)
data = {
    'track_name': [f'Song_{i}' for i in range(5000)] + ["I Don't Care (with Justin Bieber) - Loud Luxury Remix"],
    'track_artist': [f'Artist_{i % 100}' for i in range(5001)],
    'playlist_genre': ['pop', 'rock', 'hip-hop', 'electronic', 'jazz'] * 1000 + ['pop'],
    'danceability': np.random.rand(5001) * 0.8 + 0.2,
    'energy': np.random.rand(5001) * 0.8 + 0.2,
    'loudness': np.random.rand(5001) * 20 - 40,
    'speechiness': np.random.rand(5001) * 0.2,
    'acousticness': np.random.rand(5001) * 0.5,
    'instrumentalness': np.random.rand(5001) * 0.5,
    'liveness': np.random.rand(5001) * 0.4,
    'valence': np.random.rand(5001) * 0.8 + 0.2,
    'tempo': np.random.rand(5001) * 100 + 80
}
df1 = pd.DataFrame(data)

# --- Step 2: Sample Creation (ensuring the target song is included) ---
# Define the song you want to recommend
song_to_recommend = "I Don't Care (with Justin Bieber) - Loud Luxury Remix"

# Find the row corresponding to the song
song_row_df = df1[df1['track_name'] == song_to_recommend]

# Create a sample of the data that guarantees the song is included
# Take a random sample of songs *excluding* the target song.
other_songs_df = df1[df1['track_name'] != song_to_recommend].sample(n=4999, random_state=42)

# Concatenate the two DataFrames. This is the most reliable way.
sample_df = pd.concat([other_songs_df, song_row_df], ignore_index=True)

# --- Step 3: Normalization and Similarity Calculation ---
# Select the audio features for the recommendation model.
features_for_recomm = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                       'instrumentalness', 'liveness', 'valence', 'tempo']

# Normalize the features using MinMaxScaler on the sampled data
scaler = MinMaxScaler()
normalized_sample_features = scaler.fit_transform(sample_df[features_for_recomm])
cosine_sim = cosine_similarity(normalized_sample_features)

# Create a Series that maps song titles to their new index in the sampled DataFrame
indices = pd.Series(range(len(sample_df)), index=sample_df['track_name']).drop_duplicates()

# --- Step 4: Recommendation Function ---
def get_recommendations(track_name, cosine_sim_matrix=cosine_sim, df_data=sample_df, indices_map=indices):
    """
    Generates song recommendations based on content similarity.
    """
    if track_name not in indices_map.index:
        return f"'{track_name}' not found in the sampled dataset."

    idx = indices_map[track_name]
    sim_scores = list(enumerate(cosine_sim_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the top 10 most similar songs, skipping the first one (the song itself)
    sim_scores = sim_scores[1:11]

    song_indices = [i[0] for i in sim_scores]

    recommendations_df = df_data.iloc[song_indices][['track_name', 'track_artist', 'playlist_genre']]

    return recommendations_df

# --- Step 5: Example Usage and Output ---
# Run the recommendation function with the guaranteed song
recommendations = get_recommendations(song_to_recommend)

print(f"--- Top 10 recommendations for '{song_to_recommend}' ---")
print(recommendations)
```
